–ó–∞–¥–∞–Ω–∏–µ 1. –ò–º–µ–µ—Ç—Å—è —Å–ø–∏—Å–æ–∫ –∏–∑ 20 —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –æ—Ç 0 –¥–æ 100. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞–∑–±–∏–≤–∞–µ—Ç —ç—Ç–æ—Ç —Å–ø–∏—Å–æ–∫ –Ω–∞  ùëõ
  —Å–ø–∏—Å–∫–æ–≤ –ø–æ  ùëò
  —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏ –≤–∫–ª—é—á–∞–µ—Ç –≤—Å–µ —ç—Ç–∏ —Å–ø–∏—Å–∫–∏ –≤ –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ  ùëò
  –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ —Ñ—É–Ω–∫—Ü–∏–∏

````python
from random import randint

k = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–∫–º–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –≤ —Å–ø–∏—Å–∫–∞—Ö"))

def razdel_spiska(list, k):
    elem = len(list)
    n = elem // k
    res = []
    for i in range(n):
        start = i * k
        end = (i + 1) * k
        podlist = list[start:end]
        res.append(podlist)
    return res

spisok = []

for i in range(20):
    spisok.append(randint(0, 100))
print(spisok)
res = razdel_spiska(spisok, k)
print(res)
````

–ó–∞–¥–∞–Ω–∏–µ 2. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –≤—ã—á–∏—Å–ª—è—é—â—É—é —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞, –ø–µ—Ä–µ–¥–∞–≤–∞–µ–º–æ–≥–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞

````python
def factorial(n):
  fact =1
  for i in range(2, n+1):
    fact *= i

  return fact

factorial(3)
````

–ó–∞–¥–∞–Ω–∏–µ 3. –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —É–º–Ω–æ–∂–µ–Ω–∏—è, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ —Ç—Ä–µ—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤. –§—É–Ω–∫—Ü–∏—é –≤—ã–∑–≤–∞—Ç—å —Å –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–º–∏ –Ω–∏–∂–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏. –î–ª—è —Å–ª—É—á–∞—è a4 –≤—ã–±—Ä–∞—Ç—å 3 –ø–µ—Ä–≤—ã—Ö –∏ 3 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è.

a1 = (15, 10 ,5)
a2 = (3, 1)
a3 = [2, 35, 55]
a4 = (5, 10 ,15, 20)

````python
import math

def umnog(a, b=1, c=1):
    return a*b*c
   
a1 = (15,10,5)
a2=(3,1)
a3=[2,35,55]
a4=(5,10,15,20)

res1 = umnog(*a1)
res2 = umnog(*a2)
res3 = umnog(*a3)
res4 = umnog(*a4[:3])


print(res1)
print(res2)
print(res3)
print(res4)
````

–ó–∞–¥–∞–Ω–∏–µ 4. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–¥–∏—É—Å–∞ r –≤—ã—á–∏—Å–ª—è–µ—Ç –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ—Ä—Ç–µ–∂ –∏–∑ 2 –∑–Ω–∞—á–µ–Ω–∏–π

````python
import math

def ploshad_and_dlina(r):
    list = [round(math.pi*(r**2),2), round(2*math.pi*r,2)]
    return list

radius = int(input("–í–≤–µ–¥–∏—Ç–µ —Ä–∞–¥–∏—É—Å: "))
ploshad_and_dlina(radius)
````

–ó–∞–¥–∞–Ω–∏–µ 5. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —á–µ—Ç—ã—Ä–µ—Ö –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π, –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—â–∏—Ö —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –≤ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ. –°–æ–∑–¥–∞—Ç—å —Å–ª–æ–≤–∞—Ä—å —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –∏–º —Å–∏–º–≤ –æ–ª–∞–º–∏ –æ–ø–µ—Ä–∞—Ü–∏–π. –î–ª—è –¥–≤—É—Ö –∑–∞—Ä–∞–Ω–µ–µ –∑–∞–¥–∞–Ω–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª (–Ω–∞–ø—Ä–∏–º–µ—Ä, 25 –∏ 4) –≤—ã–ø–æ–ª–Ω–∏—Ç—å –≤—ã–±—Ä–∞–Ω–Ω—É—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –æ–ø–µ—Ä–∞—Ü–∏—é

````python
def operations(match, slovar):
    if match in slovar:
        return slovar[match]
    return "–ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å"

slovar = {
    "multiplication": lambda x, y: x*y,
    "plus": lambda x,y: x+y,
    "minus": lambda x,y: x-y,
    "division": lambda x, y: x//y
}

plus = operations('plus', slovar)
minus = operations('minus', slovar)
umnogenie = operations('multiplication', slovar)
delenie = operations('division', slovar)

print(plus(25,4))
print(minus(25,4))
print(umnogenie(25,4))
print(delenie(25,4))
````

–ó–∞–¥–∞–Ω–∏–µ 6. –ù–∞–ø–∏—Å–∞—Ç—å –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä –¥–ª—è —Å—Ç—Ä–æ–∫–æ–≤—ã—Ö –≤—ã—Ä–∞–∂–µ–Ω–∏–π –≤–∏–¥–∞ '<—á–∏—Å–ª–æ> <–æ–ø–µ—Ä–∞—Ü–∏—è> <—á–∏—Å–ª–æ>', –≥–¥–µ <—á–∏—Å–ª–æ> - —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –Ω–∞–ø—Ä–∏–º–µ—Ä 113, <–∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è> - –æ–¥–Ω–∞ –∏–∑ –æ–ø–µ—Ä–∞—Ü–∏–π +,-,*,//,%,^. –ü—Ä–∏–º–µ—Ä: calc('13 - 5') -> 8

````python
def calc(data):
    try:
        res = eval(data)
        return res
    except data as e:
        return "–ù–µ–≤–µ—Ä–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç"
    
data = '25-4'
result = calc(data) 
print(result) 
````

–ó–∞–¥–∞–Ω–∏–µ 8. –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (–æ—Ç 0 –¥–æ 999) –∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ –≤ —á–∏—Å–ª–æ —Ç–∏–ø–∞ int. –ò —Å–æ–æ–±—â–∞–µ—Ç –æ–± –æ—à–∏–±–∫–∞—Ö (–≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –æ–ø–∏—Å–∞–Ω–∏–µ —Ç–∏–ø–∞ –æ—à–∏–±–∫–∏ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ -1). –ü—Ä–∏–º–µ—Ä: to_int('—Ç—Ä–∏–¥—Ü–∞—Ç—å —Ç—Ä–∏') -> 33 –ü—Ä–∏–º–µ—Ä: to_int("—Å—Ç–æ —Å–æ—Ä–æ–∫ —Ç—Ä–∏–¥—Ü–∞—Ç—å –¥–≤–∞ —Å—Ç–æ") -> -1 –í—ã–≤–æ–¥: —Ç—Ä–∏–¥—Ü–∞—Ç—å - –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ –≤ —á–∏—Å–ª–µ.

````python
s1 = {
        '–Ω–æ–ª—å': 0, '–æ–¥–∏–Ω': 1, '–¥–≤–∞': 2, '—Ç—Ä–∏': 3, '—á–µ—Ç—ã—Ä–µ': 4, '–ø—è—Ç—å': 5, '—à–µ—Å—Ç—å': 6, '—Å–µ–º—å': 7, '–≤–æ—Å–µ–º—å': 8, '–¥–µ–≤—è—Ç—å': 9,
        '–¥–µ—Å—è—Ç—å': 10, '–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å': 11, '–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å': 12, '—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å': 13, '—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å': 14, '–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 15, 
        '—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å': 16, '—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 17, '–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 18, '–¥–µ–≤—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 19,
        '–¥–≤–∞–¥—Ü–∞—Ç—å': 20, '—Ç—Ä–∏–¥—Ü–∞—Ç—å': 30, '—Å–æ—Ä–æ–∫': 40, '–ø—è—Ç—å–¥–µ—Å—è—Ç': 50, '—à–µ—Å—Ç—å–¥–µ—Å—è—Ç': 60, '—Å–µ–º—å–¥–µ—Å—è—Ç': 70, '–≤–æ—Å–µ–º—å–¥–µ—Å—è—Ç': 80, '–¥–µ–≤—è–Ω–æ—Å—Ç–æ': 90,
        '—Å—Ç–æ': 100, '–¥–≤–µ—Å—Ç–∏': 200, '—Ç—Ä–∏—Å—Ç–∞': 300, '—á–µ—Ç—ã—Ä–µ—Å—Ç–∞': 400, '–ø—è—Ç—å—Å–æ—Ç': 500, '—à–µ—Å—Ç—å—Å–æ—Ç': 600, '—Å–µ–º—å—Å–æ—Ç': 700, '–≤–æ—Å–µ–º—å—Å–æ—Ç': 800, '–¥–µ–≤—è—Ç—å—Å–æ—Ç': 900
    }

def to_int(text_number):

    words = text_number.split()
    result = 0
    current_number = 0
    last_number = 0

    for word in words:
            if current_number >= s1[word]:
                if last_number > current_number:
                    print(f"{word} - –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ –≤ —á–∏—Å–ª–µ –≤ —Å—Ç—Ä–æ–∫–µ: {number_text}")
                    return -1
                last_number = current_number
                current_number = s1[word]
                result += current_number
            else:
                current_number = s1[word]
                result += current_number

    return result

number_text = '—Ç—Ä–∏–¥—Ü–∞—Ç—å —Ç—Ä–∏'
result = to_int(number_text)
print(result)

number_text = "—Å—Ç–æ —Å–æ—Ä–æ–∫ —Ç—Ä–∏–¥—Ü–∞—Ç—å –¥–≤–∞ —Å—Ç–æ"
result = to_int(number_text)
print(result)
````
