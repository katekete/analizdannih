ùëéùë•2+ùëèùë•+ùëê=0

–ó–∞–¥–∞–Ω–∏–µ 10. –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø–æ–æ—á–µ—Ä–µ–¥–Ω–æ –≤–≤–æ–¥–∏—Ç –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏ –≤ –¥–µ–∫–∞—Ä—Ç–æ–≤–æ–π —Å–∏—Å—Ç–µ–º–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –∫–∞–∫–æ–π —á–µ—Ç–≤–µ—Ä—Ç–∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –¥–∞–Ω–Ω–∞—è —Ç–æ—á–∫–∞ –∏–ª–∏ –Ω–∞ –∫–∞–∫–æ–π –æ—Å–∏ –æ–Ω–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è. –†–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ç–æ—á–∫–∏ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω. –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –Ω–æ–º–µ—Ä–∞ —á–µ—Ç–≤–µ—Ä—Ç–∏ –Ω–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ—Ç —ç—Ç–æ–π —Ç–æ—á–∫–∏ –¥–æ –Ω–∞—á–∞–ª–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –∏ –≤—ã–≤–µ—Å—Ç–∏ –µ–≥–æ –Ω–∞ —ç–∫—Ä–∞–Ω. –ï—Å–ª–∏ —Ç–æ—á–∫–∞ –ª–µ–∂–∏—Ç –Ω–∞ –æ—Å–∏, —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –Ω–æ–º–µ—Ä —á–µ—Ç–≤–µ—Ä—Ç–∏ —Ä–∞–≤–µ–Ω 0.

````python
import math

x1, y1 = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É –ø–æ –æ—Å–∏ –•: ")), int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É –ø–æ –æ—Å–∏ –£: "))
print(f"–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏: {x1}, {y1}")
rasstoyanie = math.sqrt((x1**2)+(y1**2)) # —Ä–∞—Å—Å—á–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –æ—Ç —Ç–æ—á–∫–∏ –¥–æ –Ω–∞—á–∞–ª–∞
if x1>0 and y1>0:
    print("–¢–æ—á–∫–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ 1 —á–µ—Ç–≤–µ—Ä—Ç–∏")
    print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {1*rasstoyanie}") 
elif x1<0 and y1>0:
    print("–¢–æ—á–∫–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤–æ 2 —á–µ—Ç–≤–µ—Ä—Ç–∏")
    print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {2*rasstoyanie}")  
elif x1<0 and y1<0:
    print("–¢–æ—á–∫–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ 3 —á–µ—Ç–≤–µ—Ä—Ç–∏")
    print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {3*rasstoyanie}")  
elif x1>0 and y1<0:
    print("–¢–æ—á–∫–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ 4 —á–µ—Ç–≤–µ—Ä—Ç–∏")
    print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {4*rasstoyanie}")  
elif x1 == 0:
    print("–¢–æ—á–∫–∞ –ª–µ–∂–∏—Ç –Ω–∞ –æ—Å–∏ x")
    print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {0*rasstoyanie}")  
elif y1==0:
    print("–¢–æ—á–∫–∞ –ª–µ–∂–∏—Ç –Ω–∞ –æ—Å–∏ —É")
    print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {0*rasstoyanie}") 
````

–ó–∞–¥–∞–Ω–∏–µ 11. –í—ã–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ü–µ–ª–æ–≥–æ —á–∏—Å–ª–∞ –æ—Ç 0 –¥–æ 999 –ø—Ä–æ–ø–∏—Å—å—é. –ù–∞–ø—Ä–∏–º–µ—Ä, ¬´—Å—Ç–æ –¥–µ–≤—è–Ω–æ—Å—Ç–æ –æ–¥–∏–Ω¬ª –¥–ª—è —á–∏—Å–ª–∞ 191, ¬´–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å¬ª –¥–ª—è —á–∏—Å–ª–∞ 11.

````python
ed = {
    1: "–æ–¥–∏–Ω",
    2: "–¥–≤–∞",
    3: "—Ç—Ä–∏",
    4: "—á–µ—Ç—ã—Ä–µ",
    5: "–ø—è—Ç—å",
    6: "—à–µ—Å—Ç—å",
    7: "—Å–µ–º—å",
    8: "–≤–æ—Å–µ–º—å",
    9: "–¥–µ–≤—è—Ç—å",
    0: ''
}

des = {
    2: "–¥–≤–∞–¥—Ü–∞—Ç—å",
    3: "—Ç—Ä–∏–¥—Ü–∞—Ç—å",
    4: "—Å–æ—Ä–æ–∫",
    5: "–ø—è—Ç—å–¥–µ—Å—è—Ç",
    6: "—à–µ—Å—Ç—å–¥–µ—Å—è—Ç",
    7: "—Å–µ–º–¥–µ—Å—è—Ç",
    8: "–≤–æ—Å–µ–º–¥–µ—Å—è—Ç",
    9: "–¥–µ–≤—è–Ω–æ—Å—Ç–æ"}

sot = {
    1: "—Å—Ç–æ",
    2: "–¥–≤–µ—Å—Ç–∏",
    3: "—Ç—Ä–∏—Å—Ç–∞",
    4: "—á–µ—Ç—ã—Ä–µ—Å—Ç–∞",
    5: "–ø—è—Ç—å—Å–æ—Ç",
    6: "—à–µ—Å—Ç—å—Å–æ—Ç",
    7: "—Å–µ–º—å—Å–æ—Ç",
    8: "–≤–æ—Å–µ–º—å—Å–æ—Ç",
    9: "–¥–µ–≤—è—Ç—å—Å–æ—Ç"
}

nadsat = {
    10: "–¥–µ—Å—è—Ç—å",
    11: "–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å",
    12: "–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å",
    13: "—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å",
    14: "—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å",
    15: "–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å",
    16: "—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å",
    17: "—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å",
    18: "–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å",
    19: "–¥–≤–µ—è—Ç–Ω–∞–¥—Ü–∞—Ç—å"
}

s = input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –æ—Ç 0 –¥–æ 999 :")

p = []

def f(s):
    if len(s) > 1:
        if 9 < int(s[-2:]) < 20:
            p.append(nadsat[int(s[-2:])])
        elif int(s[-2:]) > 19:
            p.append(ed[int(s[-1])])
            p.append(des[int(s[-2])])
    else:
        if int(s[-1]) > 0:
            p.append(ed[int(s[-1])])
        else:
            p.append('–Ω–æ–ª—å')
    if len(s) == 3:
        p.append(sot[int(s[-3])])
    print(' '.join(p[::-1]))
    p.clear()
f(s)
````

–ó–∞–¥–∞–Ω–∏–µ 14. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª —Å—Ç—Ä–æ–∫—É —Å–ª–µ–¥—É—é—â–µ–≥–æ –≤–∏–¥–∞: ¬´a op b¬ª, –≥–¥–µ a –∏ b ‚Äì –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —á–∏—Å–ª–∞, –∞ ‚Äòop‚Äô –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä –∏ –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è  ¬´+,‚àí,‚àó,/,‚àó‚àó,. –í –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ —Å –ø–æ–º–æ—â—å—é —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Å—Ç—Ä–æ–∫ –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ: ¬´a + b = 3 + 2 = 5¬ª, –≥–¥–µ –±—ã–ª–∞ –ø–æ–ª—É—á–µ–Ω–∞ —Å—Ç—Ä–æ–∫–∞ ¬´3 + 2¬ª. –î–ª—è –æ–ø–µ—Ä–∞—Ü–∏–π –≤–æ–∑–≤–µ–¥–µ–Ω–∏—è –≤ —Å—Ç–µ–ø–µ–Ω—å –∏ –¥–µ–ª–µ–Ω–∏—è –ø–æ –º–æ–¥—É–ª—é –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–º–µ—Å—Ç–æ –∑–Ω–∞–∫–∞ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è. –ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —Å—Ç—Ä–æ–∫–∏ ¬´2 ** 3¬ª –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤—ã–≤–µ–¥–µ–Ω–æ ¬´a –≤ —Å—Ç–µ–ø–µ–Ω–∏ b = 2 –≤ —Å—Ç–µ–ø–µ–Ω–∏ 3 = 8¬ª.

````python
input_str = input("–í–≤–µ–¥–∏—Ç–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ (a op b): ")

# —Ä–∞–∑–±–∏–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫—É –Ω–∞ –æ–ø–µ—Ä–∞–Ω–¥—ã –∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä
tokens = input_str.split()
if len(tokens) != 3:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞.")
else:
    a = float(tokens[0])
    op = tokens[1]
    b = float(tokens[2])

    # –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞
    if op == '+':
        result = a + b
    elif op == '-':
        result = a - b
    elif op == '*':
        result = a * b
    elif op == '/':
        if b != 0:
            result = a / b
        else:
            print("–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ.")
            result = None
    elif op == '**':
        result = a ** b
    elif op == '%':
        result = a % b
    else:
        print("–ù–µ–ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä.")
        result = None

    # —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ–º
    if result is not None:
        print(f"{a} {op} {b} = {input_str} = {result}")
````

–ó–∞–¥–∞–Ω–∏–µ 15. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≤–≤–µ–¥–µ–Ω–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º —á–∏—Å–ª–∞ N, –æ—Å—É—â–µ—Å—Ç–≤–ª—è–µ—Ç –≤—ã–≤–æ–¥ N –∫–æ—à–µ–∫ –Ω–∞ —ç–∫—Ä–∞–Ω. –ü—É—Å—Ç—å 1<= N <=10. –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –∫–æ—à–∫–∏ –ø—Ä–∏ N=3 –≤—ã–≥–ª—è–¥—è—Ç –ø—Ä–∏–º–µ—Ä–Ω–æ —Ç–∞–∫
 /\__/\ /\__/\ /\__/\
 /@ @\ /@ @\ /@ @\
 (~~0~~) (~~0~~) (~~0~~)

````python
cat = '''
         /\_/\ 
         /@ @\ 
        (~~0~~)
'''
res = 0
n = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π: "))
if 1<=n<=10:
    res = cat*n
    print(res)
else:
    print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ!")
````

–ó–∞–¥–∞–Ω–∏–µ 17. –ò–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ N –≤—ã–¥–µ–ª–∏—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–æ N2 —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ã–µ 2, –º–Ω–æ–∂–µ—Å—Ç–≤–æ N3 ‚Äì –∫—Ä–∞—Ç–Ω—ã–µ 3, –º–Ω–æ–∂–µ—Å—Ç–≤–æ N 6 ‚Äì –∫—Ä–∞—Ç–Ω—ã–µ 6 (—Ç.–µ. –∫—Ä–∞—Ç–Ω—ã–µ –∏ 2 –∏ 3), –º–Ω–æ–∂–µ—Å—Ç–≤–æ N 23 ‚Äì –∫—Ä–∞—Ç–Ω—ã–µ –ª–∏–±–æ 2, –ª–∏–±–æ 3. 

````python
set_1 = set()
set_N2 = set()
set_N3 = set()
set_N6 = set()
set_N23 = set()

N = int(input("–≤–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ N: "))
for number in range(1, N+1):  #–≥–µ–Ω–µ—Ä–∞—Ü–∏—è –∫–æ–Ω—Ç–µ–π–Ω–µ—Ä–∞ —Å —á–∏—Å–ª–∞–º–∏ N
    if number%2==0:
        set_N2.add(number)
        print(set_N2)
    elif number%3==0:
        set_N3.add(number)
        print(set_N3)
    elif number%6==0:
        set_N6.add(number)
    elif N%2==0 or N%3==0:
        set_N23 = set_N2.union(set_N3)
        print(set_N23)
````

–ó–∞–¥–∞–Ω–∏–µ 18. –î–∞–Ω—ã m (m>1) —Å–ª–æ–≤. –ù–∞–π—Ç–∏ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–¥–∞–Ω–Ω–æ–π –±—É–∫–≤—ã –≤ —ç—Ç–∏—Ö —Å–ª–æ–≤–∞—Ö

````python
words = input("–í–≤–µ–¥–∏—Ç–µ —Å–ª–æ–≤–∞: ")

letter = input("–í–≤–µ–¥–∏—Ç–µ –±—É–∫–≤—É –¥–ª—è –ø–æ–∏—Å–∫–∞: ")

count = 0

for word in words:
    count += word.strip().lower().count(letter.lower())

print(f"–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±—É–∫–≤ '{letter}' –≤ —Å–ª–æ–≤–∞—Ö: {count}")
````

–ó–∞–¥–∞–Ω–∏–µ 19. –î–∞–Ω—ã m (m>1) —Å–ª–æ–≤. –í –∫–∞–∫–æ–º –∏–∑ –Ω–∏—Ö –¥–æ–ª—è (–≤ %) –∑–∞–¥–∞–Ω–Ω–æ–π –±—É–∫–≤—ã –±–æ–ª—å—à–µ.

````python
words = input("–í–≤–µ–¥–∏—Ç–µ —Å–ª–æ–≤–∞: ")
letter = input("–í–≤–µ–¥–∏—Ç–µ –±—É–∫–≤—É –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞: ")

max_word = ""
max_percentage = 0

for word in words:
    word_lower = word.strip().lower()
    count_letter = word_lower.count(letter.lower())
    total_characters = len(word_lower)
    
    
    percentage = (count_letter / total_characters) * 100
    
    
    if percentage > max_percentage:
        max_word = word
        max_percentage = percentage

print(f"–°–ª–æ–≤–æ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –¥–æ–ª–µ–π –±—É–∫–≤—ã '{letter}' ({max_percentage:.2f}%) —ç—Ç–æ: {max_word}")
````
